```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
```

```python
f = np.linspace(20, 20000, 9999)
y = 24.7 * (4.37 * f / 1000 + 1)
y = np.log2((y + f) / f)
```

```python
fig, ax = plt.subplots()
fig.set_size_inches(15, 9)
fig.set_facecolor('white')
ax.set_facecolor('white')
ax.plot(f, y)
ax.semilogx()
ax.grid(True)
ax.set_xlabel('Frequency (Hz)')
ax.set_ylabel('ERB (Octaves)')
ax.xaxis.set_major_formatter(ticker.StrMethodFormatter('{x:.0f}'))
ax.set_xticks([20, 50, 100, 200, 500, 1000, 2000, 5000, 10000, 20000])
ax.set_xlim([20, 20000])
plt.show()
```

```python
fig, ax = plt.subplots()
fig.set_size_inches(15, 9)
fig.set_facecolor('white')
ax.set_facecolor('white')

for f in 50 * 2**(np.arange(9)):
    y = 24.7 * (4.37 * f / 1000 + 1)
    ax.fill_between([f - y / 2, f + y / 2], [0, 0], [1, 1], color='C0')

ax.semilogx()
ax.set_xlabel('Frequency (Hz)')
ax.xaxis.set_major_formatter(ticker.StrMethodFormatter('{x:.0f}'))
ax.set_xticks([20, 50, 100, 200, 500, 1000, 2000, 5000, 10000, 20000])
ax.set_xlim([20, 20000])
ax.set_yticks([])
ax.set_ylim([0, 1.1])
plt.show()
```
