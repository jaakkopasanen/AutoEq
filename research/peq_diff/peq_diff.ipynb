```python
import sys
from pathlib import Path
ROOT_PATH = Path().resolve().parent.parent
if str(ROOT_PATH) not in sys.path:
    sys.path.insert(1, str(ROOT_PATH))
import numpy as np
import matplotlib.pyplot as plt
from frequency_response import FrequencyResponse
```

```python
fig, ax = plt.subplots()
fig.set_size_inches(15, 9)
for i, fp in enumerate(ROOT_PATH.joinpath('research/peq_diff').glob('*.csv')):
    fr = FrequencyResponse.read_from_csv(fp)
    fr.plot_graph(fig=fig, ax=ax, raw_plot_kwargs={'color': f'C{i}', 'label': fr.name, 'linewidth': 1.5}, show=False)
ax.legend()
ax.set_xlim([1000, 20000])
ax.set_title('Difference between parametric equalizers')
ax.set_facecolor('white')
fig.savefig('peq_diff.png')
plt.show()
```
